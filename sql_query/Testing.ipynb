```python
from pyspark.sql import functions as F, types as T
```

```python
path = "/Volumes/fmcg/gold/parent_incremental_update/fact_orders.csv"
df = spark.read.format("csv").option("header", "true").load(path)
display(df.limit(5))
```

```python
df_dim_date = spark.sql(f"select * from fmcg.gold.dim_date")
df_fact_orders = spark.sql(f"select * from fmcg.gold.fact_orders")
df_joined = df_fact_orders.join(
    df_dim_date, on=df_fact_orders.date == df_dim_date.month_start_date, how="left"
)
display(df_joined.limit(5))
df_joined.count()
```
